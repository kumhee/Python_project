```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./data/PolynomialRegressionData.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X,y
```

```python
# 선형 회귀 모델 생성
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
```

```python
plt.figure(figsize=(8, 5), dpi=80)
plt.scatter(X, y, color='lightpink')
plt.plot(X, reg.predict(X), color='purple')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.score(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) #2차 방정식 객체 생성
X_poly = poly_reg.fit_transform(X)
```

```python
reg = LinearRegression()
reg.fit(X_poly, y)
```

```python
plt.figure(figsize=(8, 5), dpi=80)
plt.scatter(X, y, color='blue')
plt.plot(X, reg.predict(poly_reg.fit_transform(X)), color='green')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
# 곡선을 부드럽게 출력하기 위해 독립변수(X)을 일정하게 증가시켜 새로운 값을 생성해 준다. (Numpy 이용)
X_range = np.arange(min(X), max(X), 0.1) #X축을 최솟값~최댓값 0.1단위로 잘라서 데이터를 생성
X_range
```

```python
X_range=X_range.reshape(-1, 1) #row은 전체 데이터 수: len(X_range) 또는 -1, column의 1로 지정
X_range.shape
```

```python
plt.figure(figsize=(8, 5), dpi=80)
plt.scatter(X, y, color='purple')
plt.plot(X_range, reg.predict(poly_reg.fit_transform(X_range)), color='lightpink')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
# 데이터 읽기
df = pd.read_csv('./data/PolynomialRegressionData.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# X독립변수를 다항으로 변경
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) #2차 방정식 객체 생성
X_poly = poly_reg.fit_transform(X)
```

```python
# 선행객체 만들어서 학습
reg = LinearRegression()
reg.fit(X_poly, y)
```

```python
# X 범위를 0.1단위로 생성
X_range = np.arange(min(X), max(X), 0.1)
X_range=X_range.reshape(-1, 1) 
```

```python
# 그래프 출력
plt.figure(figsize=(8, 5), dpi=80)
plt.scatter(X, y, color='purple')
plt.plot(X_range, reg.predict(poly_reg.fit_transform(X_range)), color='lightpink')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```
